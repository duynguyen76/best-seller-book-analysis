# **Import all the needed library**




```python
# Data handling
import pandas as pd
import numpy as np
```

# **Import the data from csv file**

The dataset is in CSV format; hence we use the `read_csv` function from Pandas.  
Immediately after importing, take a quick look at the data using the `info` and `head` method.

```python
#Read csv file
book_data = pd.read_csv("kindle_data.csv")
book_data.info()
book_data.head()
```

# **Clean the data**

Before analysing the data, a good practice is to clean the data, including deleting Null variables and redundant columns as well as checking for duplicates.

### **I. Remove the blank box in "author" column**
From the information we got from `info` method, we know that **"author"** column have some null values. As we will use the information from this column later, we will drop all data with blank author.

```python
#Remove row blank author
book_data_cleaned = book_data.dropna(subset=['author'])
book_data_cleaned.info()
book_data_cleaned.head()
```

### **II. Remove unnecessary columns**
Our DataFrame contains 16 columns, but we would not use all of them. To make it cleaner, easier to view and less complex, we will remove the unnecessary columns, such as **"soldBy", "imgUrl", "productURL", "reviews"** and **"publishedDate"**.

```python
#Remove unnecessary columns
columns_to_remove = ["soldBy", "imgUrl", "productURL", "reviews","publishedDate"]
book_data_cleaned = book_data_cleaned.drop(columns=columns_to_remove)
book_data_cleaned.info()
book_data_cleaned.head()
```

### **III. Check for duplicate books**
Now we will check for duplicate books, by checking number of duplicated **"asin"**, which is unique for every book. If there is any duplicated value, we will keep the first data and delete the others.

```python
#Check for duplicate books using asin
duplicate_count = book_data_cleaned.duplicated(subset=['asin']).sum()

if duplicate_count > 0:
    book_data_cleaned = book_data_cleaned.drop_duplicates(subset=['asin'], keep='first')
    print(f"Removed {duplicate_count} duplicate ASINs. File updated successfully.")

else:
    print("No duplicate ASINs found. The file remains the same.")
```

### **IV. Remove wrong asin values**
For kindle books, asin is a string of characters and numbers instead of 10-digit ISBN, which is for physical books. Therefore, we will remove all books with integer asin.

```python
#drop asin that is int (wrong format)
book_data_cleaned = book_data_cleaned[~book_data_cleaned['asin'].astype(str).str.isdigit()]
book_data_cleaned.info()
```

### **V. Save book_data_cleaned into csv file**
After cleaning the data, we will save it into a new csv file for later analysis.

```python
book_data_cleaned.to_csv('book_data_cleaned.csv', index=False)
```
